### Surgery List + TIOBIO

```python
import pandas as pd 

# Read the data
# data = pd.read_csv(r'C:\Users\e745092\Desktop\TIOB\IO Denominator Misc_Reports\surgery_detail_tiob_cases.csv')

# SurgeryFY21-FY23.csv -- old data included with new csv file
data = pd.read_csv(r'C:\Users\e745092\Desktop\Surgery+TIOB\Surgery Detail csv\SurgeryFY21-FY23.csv')
data.head()
```

```python
# Drop row 0 and 1 using index number

# data.drop([0,1], inplace=True)

# # rename the header using the first row

# data.columns = data.iloc[0]

# # Drop the first row

# data.drop([2], inplace=True)

# select columns

surgery_tiob_cases = data[[
 'DIV',
 'DB DESC',
 'FY',
 'TIOB Group',
 'PATIENT ID',
 'MRN',
 'ADMIT DATE',
 'PRIM DX',
 'OR Procedure Name',
 'Primary Surgeon Name',]]


```

```python
# drop null values in MRN column
surgery_tiob_cases = surgery_tiob_cases.dropna(subset=['MRN'])


surgery_tiob_cases['MRN'] = surgery_tiob_cases['MRN'].astype(int)

surgery_tiob_cases
```

### Bring in TIOB IO

```python
# TIOB io 780 rows
tiob_io = pd.read_csv(r'C:\Users\e745092\Desktop\Shraddha Email TIOB IO\TIOB_IO_07052023.csv')
# tiob_io = pd.read_csv(r'C:\Users\e745092\Desktop\io_tiob_updated.csv')

tiob_io.info()
```

```python
# change column name _merge to From Report 
# tiob_io.rename(columns={'_merge':'From Report'}, inplace=True)
```

```python
# inner join tiob_io and surgery_tiob_cases

surgery_tiob_inner = pd.merge(surgery_tiob_cases, tiob_io, how='inner', on='MRN', indicator=True)
surgery_tiob_inner.info() 
```

```python
# to csv
# surgery_tiob_inner.to_csv(r'C:\Users\e745092\Desktop\surgery_tiob_inner_matching.csv', index=False)
```

```python
# outer merge tiob_io and surgery_tiob_cases

surgery_tiob_outer = pd.merge(surgery_tiob_cases, tiob_io, how='outer', on='MRN', indicator=True)
surgery_tiob_outer.info()
```

```python
# append both 

df_append = surgery_tiob_inner.append(surgery_tiob_outer)
```

### LOCATIONS 


```python
# surgery locations db desc 

surgery_tiob_cases['DB DESC'].unique().tolist()
```

```python
df_append['LOCATIONS'] = df_append['Site'].fillna(df_append['DB DESC'])
```

```python
df_append.info()
```

```python
# location name updates 


df_append['LOCATIONS'].replace(to_replace='COLLEGE STATION', value='College Station', inplace=True)
df_append['LOCATIONS'].replace(to_replace='ROUND ROCK', value='Round Rock', regex=True, inplace=True)
df_append['LOCATIONS'].replace(to_replace='TEMPLE', value='Temple', regex=True, inplace=True)
df_append['LOCATIONS'].replace(to_replace='WACO HILLCREST', value='Waco', regex=True, inplace=True)
df_append['LOCATIONS'].replace(to_replace='ALL SAINTS', value='BAS', regex=True, inplace=True)
df_append['LOCATIONS'].replace(to_replace='PLANO', value='Plano', regex=True, inplace=True)
df_append['LOCATIONS'].replace(to_replace='WAXAHACHIE', value='WAX/DUN', regex=True, inplace=True)

```

```python
df_append['LOCATIONS'].unique().tolist()
```

# DATE COLUMN 
use consent date and fill na with admit date 

```python
df_append['ADMIT CONSENT DATE'] = df_append['Consent Date'].fillna(df_append['ADMIT DATE'])
```
